```python
import pandas as pd
```

```python
products = pd.read_csv("product_info.csv")
products.head()
```

```python
products.info()
products.isnull().sum()
```

```python
products.describe(include='all')
```

```python
import glob

review_files = glob.glob("reviews_*.csv")
review_files
```

```python
reviews_list = []

for file in review_files:
    df = pd.read_csv(file, engine='python')
    reviews_list.append(df)
```

```python
reviews = pd.concat(reviews_list, ignore_index=True)
reviews.head()
```

```python
reviews.shape
```

```python
reviews.isnull().sum()
```

```python
products = products.drop_duplicates(subset="product_id")
products.shape
```

```python
products.columns
```

```python
products["price_usd"].head()
```

```python
products["price_usd"].isnull().sum()
```

```python
products["reviews"] = products["reviews"].fillna(0).astype(int)
```

```python
products["value_price_usd"] = products["value_price_usd"].fillna(0)
```

```python
products["sale_price_usd"] = products["sale_price_usd"].fillna(products['price_usd'])
```

```python
products["secondary_category"] = products["secondary_category"].fillna("Unknown")
products["tertiary_category"] = products["tertiary_category"].fillna("Unknown")
```

```python
reviews["is_recommended"] = reviews["is_recommended"].fillna(0).astype(int)
```

```python
reviews = reviews.dropna(subset=["review_text"])
```

```python
user_cols = ["skin_tone", "eye_color", "skin_type", "hair_color"]
reviews[user_cols] = reviews[user_cols].fillna("Unknown")
```

```python
brand_map = products.set_index("product_id")["brand_name"]
price_map = products.set_index("product_id")["price_usd"]

reviews["brand_name"] = reviews["brand_name"].fillna(
    reviews["product_id"].map(brand_map)
)

reviews["price_usd"] = reviews["price_usd"].fillna(
    reviews["product_id"].map(price_map)
)
```

```python
reviews.isnull().sum()
```

```python
products.isnull().sum()
```

```python
import numpy as np
```

```python
products[['price_usd', 'sale_price_usd']].head(10)
```

```python
products["effective_price_usd"] = products["sale_price_usd"].fillna(
    products["price_usd"]
)
```

```python
products['discount_percentage'] = np.where(products['price_usd'] > 0,
 ((products['price_usd'] - products['effective_price_usd'] )/ products['price_usd']) * 100,
 0)
```

```python
products['discount_percentage'] = products['discount_percentage'].round(2)
```

```python
products['discount_percentage'].describe()
```

```python
products.loc[products["discount_percentage"] > 0, "discount_percentage"].describe()
```

```python
products["discount_bucket"] = pd.cut(
    products["discount_percentage"],
    bins=[-0.1, 0.01, 10, 20, 30, 50, 100],
    labels=["No Discount", "<10%", "10–20%", "20–30%", "30–50%", "50%+"],
    right=True
)
```

```python
products['discount_bucket'].head()
```

```python
products[products["discount_percentage"] > 70][
    ["brand_name", "product_name", "price_usd", "sale_price_usd", "discount_percentage"]
].head()
```

```python
products["on_sale"] = (products["discount_percentage"] > 0).astype(int)
```

```python
products["on_sale"].head()
```

```python
labels = ['Low', 'Medium', 'High']
products['popularity'] = pd.qcut(products['loves_count'], q=3, labels=labels)
```

```python
products[['loves_count', 'popularity']].head(20)
```

```python
reviews['rating'].head()
```

```python
labels = ['Underperforming', 'Satisfactory', 'High Rated']
products['rating_category'] = pd.qcut(products['rating'], q=3, labels=labels)
```

```python
products[['popularity','rating_category']].head()
```

```python
products["price_range"] = pd.cut(
    products["price_usd"],
    bins=[0, 25, 50, 100, 200, 1000],
    labels=["<25", "25-50", "50-100", "100-200", "200+"]
)
```

```python
products[['price_usd', 'price_range']].head(10)
```

```python
products["has_reviews"] = (products["reviews"] > 0).astype(int)
```

```python
products[['reviews', 'has_reviews']].head()
```
